## Now let's take a stab at training a linear regression. (FUTURE: we can use same training data for any type of Machine Learning algorithm ie. SVM, Neural Network)

### let's make training data from 1D and 4D separately to compare

```python
from sklearn.model_selection import train_test_split
ind_startx = 'w'
ind_endx = 'w'
ind_y = 'distance'
# first ALL data as numpy arrays
X = df_clean.loc[:, ind_startx:ind_endx].values
y = df_clean.loc[:, ind_y].values
X_train_1d, X_test_1d, y_train_1d, y_test_1d = train_test_split(X, y, test_size=0.30, random_state=42)
# now split data into train and test set
print('we have {} train, {} test data'.format(len(X_train_1d), len(X_test_1d)))
```

```python
# from sklearn.model_selection import train_test_split
ind_startx_4d = 'u'
ind_endx_4d = 'h'
ind_y_4d = 'distance'
# first ALL data as numpy arrays
X_4d = df_clean.loc[:, ind_startx_4d:ind_endx_4d].values
y_4d = df_clean.loc[:, ind_y_4d].values
X_train_4d, X_test_4d, y_train_4d, y_test_4d = train_test_split(X_4d, y_4d, test_size=0.30, random_state=42)
# now split data into train and test set
print('we have {} train, {} test data of ndims {}'.format(len(X_train_4d), len(X_test_4d), X_4d.shape[1]))
```

```python
def make_train_test_data_4d(dataframe):
    ind_startx_4d = 'u'
    ind_endx_4d = 'h'
    ind_y_4d = 'distance'
    # first ALL data as numpy arrays
    X_4d = dataframe.loc[:, ind_startx_4d:ind_endx_4d].values
    y_4d = dataframe.loc[:, ind_y_4d].values
    return train_test_split(X_4d, y_4d, test_size=0.30, random_state=42)

def train_test(model, X_train, y_train, X_test, y_test):
    # Train the model using the training sets
    model.fit(X_train, y_train)
    # Make predictions using the testing set
    y_pred = model.predict(X_test)
    results_df = pd.DataFrame(columns=['y_groundtruth', 'y_predicted', 'error', 'error_percent'])
    results_df.y_groundtruth = pd.Series(data=y_test)
    results_df.y_predicted = pd.Series(data=y_pred)
    results_df.error = np.abs(results_df.y_groundtruth - results_df.y_predicted)
    results_df.error_percent = results_df.error / results_df.y_groundtruth
    return results_df

def plot_predictions(dataframe, start_idx=0, end_idx=20):
    %matplotlib inline
    x_axis = dataframe.index
    y_axis_gt = dataframe.y_groundtruth
    y_axis_pred = dataframe.y_predicted
    plt.plot(x_axis[start_idx:end_idx], y_axis_gt[start_idx:end_idx], c='b', label='ground truth') # plot the ground truth of test data
    plt.plot(x_axis[start_idx:end_idx], y_axis_pred[start_idx:end_idx], c='r', label='predicted') # plot the predicted
    plt.xlabel('data point')
    plt.ylabel('distance in ft')
    plt.legend()
    # plt.scatter(X_train, y_train, c='y')
    # plt.scatter(X_train, y_pred_t, c='g')
    plt.show()

def plot_errors(dataframe, start_idx=0, end_idx=20):
    %matplotlib inline
    x_axis = dataframe.index
    error_1 = dataframe.error
    plt.plot(x_axis[start_idx:end_idx], error_1[start_idx:end_idx], c='b', label='error') # plot the ground truth of test data
    plt.xlabel('data point')
    plt.ylabel('error')
    plt.legend()
    # plt.scatter(X_train, y_train, c='y')
    # plt.scatter(X_train, y_pred_t, c='g')
    plt.show()    
    
def compare_errors(dataframe_1, dataframe_2, start_idx=0, end_idx=20):
    %matplotlib inline
    x_axis = dataframe_1.index
    error_1 = dataframe_1.error
    error_2 = dataframe_2.error
    plt.plot(x_axis[start_idx:end_idx], error_1[start_idx:end_idx], c='b', label='error_1') # plot the ground truth of test data
    plt.plot(x_axis[start_idx:end_idx], error_2[start_idx:end_idx], c='r', label='error_2') # plot the predicted
    plt.xlabel('data point')
    plt.ylabel('error')
    plt.legend()
    # plt.scatter(X_train, y_train, c='y')
    # plt.scatter(X_train, y_pred_t, c='g')
    plt.show()
```

# Compare 1D vs. 4D Linear Regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR

linear_regression = LinearRegression()
# regr_svr_4d = SVR(kernel='rbf')
results_1d = train_test(model=linear_regression, 
                        X_train=X_train_1d, 
                        y_train=y_train_1d, 
                        X_test=X_test_1d, 
                        y_test=y_test_1d)

results_4d = train_test(model=linear_regression, 
                        X_train=X_train_4d, 
                        y_train=y_train_4d, 
                        X_test=X_test_4d, 
                        y_test=y_test_4d)
compare_errors(dataframe_1=results_1d, dataframe_2=results_4d)
```

# Compare SVR vs. Linear Regression

```python
def compare_2_models(model1, 
                     model2,
                     X_train, 
                     y_train, 
                     X_test, 
                     y_test, ):
    
    results_1 = train_test(model=model1, 
                            X_train=X_train, 
                            y_train=y_train, 
                            X_test=X_test, 
                            y_test=y_test)
    results_2 = train_test(model=model2, 
                        X_train=X_train, 
                        y_train=y_train, 
                        X_test=X_test, 
                        y_test=y_test)


    compare_errors(dataframe_1=results_1, dataframe_2=results_2)
```

## The SVR seems to have trouble with shorter distances, perhaps we need to experiment with the SVR parameters

```python
# let's sweep the C value to see if that helps
import numpy as np
c_vals = np.arange(0.1, 20.0, 0.5)

for c in c_vals:
    svr = SVR(kernel='rbf', C=c)
    results_lr = train_test(model=linear_regression, 
                            X_train=X_train_4d, 
                            y_train=y_train_4d, 
                            X_test=X_test_4d, 
                            y_test=y_test_4d)

    results_svr = train_test(model=svr, 
                            X_train=X_train_4d, 
                            y_train=y_train_4d, 
                            X_test=X_test_4d, 
                            y_test=y_test_4d)
#     plt.title('C={}'.format(c))
#     compare_errors(dataframe_1=results_lr, dataframe_2=results_svr)
    plt.title('C={}'.format(c))
    plot_predictions(dataframe=results_svr)
```

## Looks like as C increases, the more the model tends to fit to the data. We need to beware of overfitting, but it looks like roughly C=5 to 10 is what we want

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
C = 20
linear_regression = LinearRegression()
svr = SVR(kernel='rbf', C=C)
compare_2_models(model1=linear_regression, 
                 model2=svr,
                 X_train=X_train_4d, 
                 y_train=y_train_4d, 
                 X_test=X_test_4d, 
                 y_test=y_test_4d)
```

```python
import os
import glob
import pandas as pd
def extract_bb_lidar_from_csv(csv_file):
    df = pd.DataFrame.from_csv(csv_file)
    bboxes_detected = []
    lidar_distances = []
    if 'Box Detected' in df.columns and 'Distance Lidar' in df.columns:
        for box_detected, lidar_distance in zip(df['Box Detected'], df['Distance Lidar']):
            bbox = [int(s) for s in box_detected.replace('[','').replace(']','').replace(',','').split(' ') if len(s)>0]
            bboxes_detected.append(np.array(bbox))
            lidar_distances.append(float(lidar_distance))
        bb_series = pd.Series(data=bboxes_detected)
        lidar_series = pd.Series(data=lidar_distances)
        output_df = pd.DataFrame(columns=['bbox', 'lidar'])
        output_df['bbox'] = bb_series
        output_df['lidar'] = lidar_series
        return output_df

def clean_data(dataframe):
    """
    remove bbox [-1, -1, -1, -1] entries
    remove lidar 0.0 entries
    """
    is_good = []
    for bb, dist in zip(dataframe['bbox'], dataframe['lidar']) :
        if bb is not None and dist is not None:
            if (bb == np.array([-1,-1,-1,-1])).all() or not dist > 0:
                is_good.append(False)
            else:
                is_good.append(True)
        else:
            is_good.append(False)
    return dataframe[is_good].copy()

def normalize_data(dataframe):
    pass
```

### Looks like the Johnson SU fits best, let's transform to that!

```python
def johnson(y):
    gamma, eta, epsilon, lbda = st.johnsonsu.fit(y)
    yt = gamma + eta*np.arcsinh((y-epsilon)/lbda)
    return yt, gamma, eta, epsilon, lbda
def johnson_inverse(y, gamma, eta, epsilon, lbda):
    return lbda*np.sinh((y-gamma)/eta) + epsilon
yt, g, et, ep, l = johnson(df_train['w'].copy())
yt2 = johnson_inverse(yt, g, et, ep, l)
df_transformed = pd.DataFrame(columns=['w'],data=yt) # TODO: do necessary transforms for other columns
# plt.subplot(131)
plt.title('before transformation')
sns.distplot(df_train['w'], bins=100, fit=st.norm, kde=False)
plt.show()
# plt.subplot(132)
plt.title('after transformation')
sns.distplot(df_transformed['w'], bins=100, fit=st.norm, kde=False)
plt.show()
# plt.subplot(133)
plt.title('inverse transformation')
sns.distplot(yt2, bins=100, fit=st.norm, kde=False)
plt.show()
```
